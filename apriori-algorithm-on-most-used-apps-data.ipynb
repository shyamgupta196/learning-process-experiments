what is association rule mining 

Association rule mining finds interesting associations and relationships among large sets of data items. This rule shows how frequently a itemset occurs in a transaction. 

Before we start defining the rule, let us first see the basic definitions.

* Support Count(sigma) – Frequency of occurrence of a itemset.

* Frequent Itemset – An itemset whose support is greater than or equal to minsup threshold.

* Association Rule – An implication expression of the form X -> Y, where X and Y are any 2 itemsets.

* Support(s) –
    The number of transactions that include items in the {X} and {Y} parts of the rule as a percentage of the total number of transaction.It is a measure of how frequently the collection of items occur together as a percentage of all transactions.

*    **FORMULA** support = sigma(X+Y)/total items –
        It is interpreted as fraction of transactions that contain both X and Y.
* Confidence(c) –
    It is the ratio of the no of transactions that includes all items in {B} as well as the no of transactions that includes all items in {A} to the no of transactions that includes all items in {A}.
    
    formula **Conf(X=>Y) = Supp(XUY) / Supp(X)** –
    It measures how often each item in Y appears in transactions that contains items in X also.
*    Lift(l) –
        The lift of the rule X=>Y is the confidence of the rule divided by the expected confidence, assuming that the itemsets X and Y are independent of each other.The expected confidence is the confidence divided by the frequency of {Y}.
* Lift(X=>Y) = Conf(X=>Y) / Supp(Y) –
    Lift value near 1 indicates X and Y almost often appear together as expected, greater than 1 means they appear together more than expected and less than 1 means they appear less than expected.Greater lift values indicate stronger association.


```python
te = TransactionEncoder()
te_data = te.fit(df).transform(df)
data = pd.DataFrame(te_data,columns=te.columns_)
data
```

```python
freq_items = apriori(data, min_support = 0.10, use_colnames = True, verbose = 1)
freq_items.sort_values("support", ascending = False).head(20)
```

```python
### interpretation :--
* youtube alone 
```

```python
freq_items.sort_values('support',ascending=False).tail(10)
```

```python
# the support values for the head n tail of the data 
# from support we can infer that the combination of n pairs(n<4 in this case) is greater than 10% support in the worst scenario ,
# which means
# there is a 10% or more out of the total data that the person will install or watch or view the consequent paired app 
# youtube and whatsapp being highest in support pairs of 2 tells us that out of 100% there is 87.0968% chance of both being downloaded or Watched 
```

```python
df_ar = association_rules(freq_items, metric = "confidence", min_threshold = 0.5)
df_ar
```

```python
df_ar[(df_ar.support > 0.15) & (df_ar.confidence > 0.5)].sort_values("confidence", ascending = False)
```

- Antecedent support variable tells us probability of antecedent products alone
- Consequents support variable tells us probability of consequents products alone
- The support value is the value of the two products (Antecedents and Consequents)
