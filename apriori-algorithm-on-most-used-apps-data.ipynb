<h3 style="color:orange">
what is association rule mining???? 

Association rule mining finds interesting associations and relationships among large sets of data items. This rule shows how frequently a itemset occurs in a transaction. 
</h3>

<p style="color:orange">
Before we start implementation of <b> # the apriori rule</b> 
    </p>
    <h2 style="color:purple">
    <b>let us first see the basic definitions.⬇⬇⬇</b>
    </h2>
    

    * Support Count(sigma) – Frequency of occurrence of a itemset.

    * Frequent Itemset – An itemset whose support is greater than or equal to minsup threshold.

    * Association Rule – An implication expression of the form X -> Y, where X and Y are any 2 itemsets.

    * Support(s) –
    The number of transactions that include items in the {X} and {Y} parts of the rule as a percentage of the total number of transaction.It is a measure of how 
    frequently the collection of items occur together as a percentage of all transactions.

    *    **FORMULA** support = sigma(X+Y)/total items –
    It is interpreted as fraction of transactions that contain both X and Y.
    * Confidence(c) –
    It is the ratio of the no of transactions that includes all items in {B} as well as the no of transactions that includes
    all items in {A} to the no of transactions that includes all items in {A}.

    formula Conf(X=>Y) = **Supp(XUY) / Supp(X)** –
    It measures how often each item in Y appears in transactions that contains items in X also.
    *    Lift(l) –
    The lift of the rule X=>Y is the confidence of the rule divided by the expected confidence, assuming that the itemsets X and Y are independent of each other.The 
    expected confidence is the confidence divided by the frequency of {Y}.
    * Lift(X=>Y) = Conf(X=>Y) / Supp(Y) –
<b>
        Lift value near 1 indicates X and Y almost often appear together as expected, greater than 1 means they appear together more than expected and less than 1 means they appear less than expected.Greater lift values indicate stronger association.</b>

</html>

```python
te = TransactionEncoder()
te_data = te.fit(df).transform(df)
data = pd.DataFrame(te_data,columns=te.columns_)
data
```

so,

to keep it simple we have used a very low support price for the analysis 

min_support == .30  i.e (30%)

```python
freq_items = apriori(data, min_support = 0.30, use_colnames = True, verbose = 1)
freq_items.sort_values("support", ascending = False)
```

```python
df_ar = association_rules(freq_items, metric = "confidence", min_threshold = 0.5)
df_ar
```
